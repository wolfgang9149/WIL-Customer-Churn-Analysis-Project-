```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Load the cleaned dataset
url = 'https://drive.google.com/uc?id=1qBkAiPPQ9bTiaY6PcDq7Tmp8DQsCyyrZ'
df = pd.read_csv(url)

# Convert categorical variables to numeric
df_encoded = pd.get_dummies(df)

# Separate features and target variable
target_column = 'Churn_No'  # Use 'Churn_No' as the target variable
X = df_encoded.drop([target_column, 'Churn_Yes'], axis=1)  # Drop target and the other dummy column
y = df_encoded[target_column]  # Target

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Convert data to DataFrames for saving
X_train_df = pd.DataFrame(X_train, columns=X.columns)
X_test_df = pd.DataFrame(X_test, columns=X.columns)
y_train_df = pd.DataFrame(y_train, columns=['Churn_No'])
y_test_df = pd.DataFrame(y_test, columns=['Churn_No'])

# Save the DataFrames to CSV files
X_train_df.to_csv('X_train_Churn_No.csv', index=False)
X_test_df.to_csv('X_test_Churn_No.csv', index=False)
y_train_df.to_csv('y_train_Churn_No.csv', index=False)
y_test_df.to_csv('y_test_Churn_No.csv', index=False)

print("Data scaling, splitting, and saving completed successfully with 'Churn_No'.")
print("Files saved: X_train_Churn_No.csv, X_test_Churn_No.csv, y_train_Churn_No.csv, y_test_Churn_No.csv")
```

```python
import pandas as pd

# Load the CSV files
X_train = pd.read_csv('X_train_Churn_No.csv')
X_test = pd.read_csv('X_test_Churn_No.csv')
y_train = pd.read_csv('y_train_Churn_No.csv')
y_test = pd.read_csv('y_test_Churn_No.csv')

# Display the first few rows of each dataset
print("X_train:")
print(X_train.head())
print("\nX_test:")
print(X_test.head())
print("\ny_train:")
print(y_train.head())
print("\ny_test:")
print(y_test.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of target variable in training set
plt.figure(figsize=(8, 6))
sns.countplot(x='Churn_No', data=y_train)
plt.title('Distribution of Churn_No in Training Set')
plt.xlabel('Churn_No')
plt.ylabel('Count')
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(X_train.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Matrix')
plt.show()
```

```python
# Combine X_train and y_train for pairplot
X_train_with_target = X_train.copy()
X_train_with_target['Churn_No'] = y_train

# Pairplot
sns.pairplot(X_train_with_target, hue='Churn_No')
plt.title('Pairplot of Features Colored by Churn_No')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

# Fit a RandomForest model
model = RandomForestClassifier()
model.fit(X_train, y_train.values.ravel())

# Feature importance
importances = model.feature_importances_
features = X_train.columns

# Plot feature importance
plt.figure(figsize=(12, 6))
sns.barplot(x=importances, y=features)
plt.title('Feature Importances')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```
